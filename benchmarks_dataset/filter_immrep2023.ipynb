### Remove the IMMREP2023 samples that appeared in the train&valid sets

```python
import os
import pandas as pd
```

level 1

```python
level = "1"
data_path = f"./level_{level}"
train_df = pd.read_csv(os.path.join(data_path, "train_data_fold{}.csv".format(0)))
valid_df = pd.read_csv(os.path.join(data_path, "valid_data_fold{}.csv".format(0)))
train_df = train_df[train_df['label']==1]
valid_df = valid_df[valid_df['label']==1]

train_df['tag'] = train_df['pep'] + '+' + train_df['beta']
valid_df['tag'] = valid_df['pep'] + '+' + valid_df['beta']

immrep_df = pd.read_csv('../data/immrep2023_solutions.csv')
immrep_df = immrep_df[immrep_df['Label']==1]
immrep_df['tag'] = immrep_df['Peptide'] + '+' + immrep_df['CDR3b']
immrep_df['tag_extended'] = immrep_df['Peptide'] + '+' + immrep_df['CDR3b_extended']
filtered_tags = list(set(immrep_df['tag'].tolist()) | set(immrep_df['tag_extended'].tolist()))

filtered_train_df = train_df[~train_df['tag'].isin(filtered_tags)]
filtered_valid_df = valid_df[~valid_df['tag'].isin(filtered_tags)]
print(f'Train positives: {len(train_df):6d} -> {len(filtered_train_df):6d} ({len(train_df)-len(filtered_train_df):4d} filtered)')
print(f'Valid positives: {len(valid_df):6d} -> {len(filtered_valid_df):6d} ({len(valid_df)-len(filtered_valid_df):4d} filtered)')

filtered_train_df = filtered_train_df.copy().drop(columns=['tag'])
filtered_valid_df = filtered_valid_df.copy().drop(columns=['tag'])
filtered_train_df.to_csv(os.path.join(data_path, "filtered_train_data_fold0.csv"), index=False)
filtered_valid_df.to_csv(os.path.join(data_path, "filtered_valid_data_fold0.csv"), index=False)
```

level 2

```python
level = "2"
data_path = f"./level_{level}"
train_df = pd.read_csv(os.path.join(data_path,"train_data_fold{}.csv".format(0)))
valid_df = pd.read_csv(os.path.join(data_path,"valid_data_fold{}.csv".format(0)))
train_df = train_df[train_df['label']==1]
valid_df = valid_df[valid_df['label']==1]

train_df['tag'] = train_df['pep'] + '+' + train_df['hla.allele'] + '+' + train_df['beta']
valid_df['tag'] = valid_df['pep'] + '+' + valid_df['hla.allele'] + '+' + valid_df['beta']

immrep_df = pd.read_csv('../data/immrep2023_solutions.csv')
immrep_df = immrep_df[immrep_df['Label']==1]
immrep_df['HLA'] = immrep_df['HLA'].apply(lambda x: 'HLA-'+x)
immrep_df['tag'] = immrep_df['Peptide'] + '+' + immrep_df['HLA'] + '+' +immrep_df['CDR3b']
immrep_df['tag_extended'] = immrep_df['Peptide'] + '+' + immrep_df['HLA'] + '+' +immrep_df['CDR3b_extended']
filtered_tags = list(set(immrep_df['tag'].tolist()) | set(immrep_df['tag_extended'].tolist()))

filtered_train_df = train_df[~train_df['tag'].isin(filtered_tags)]
filtered_valid_df = valid_df[~valid_df['tag'].isin(filtered_tags)]
print(f'Train positives: {len(train_df):6d} -> {len(filtered_train_df):6d} ({len(train_df)-len(filtered_train_df):4d} filtered)')
print(f'Valid positives: {len(valid_df):6d} -> {len(filtered_valid_df):6d} ({len(valid_df)-len(filtered_valid_df):4d} filtered)')

filtered_train_df = filtered_train_df.copy().drop(columns=['tag'])
filtered_valid_df = filtered_valid_df.copy().drop(columns=['tag'])
filtered_train_df.to_csv(os.path.join(data_path, "filtered_train_data_fold0.csv"), index=False)
filtered_valid_df.to_csv(os.path.join(data_path, "filtered_valid_data_fold0.csv"), index=False)
```

level 3

```python
level = "3"
data_path = f"./level_{level}/"
train_df = pd.read_csv(os.path.join(data_path,"train_data_fold{}.csv".format(0)))
valid_df = pd.read_csv(os.path.join(data_path,"valid_data_fold{}.csv".format(0)))
train_df = train_df[train_df['label']==1]
valid_df = valid_df[valid_df['label']==1]

train_df['tag'] = train_df['pep'] + '+' + train_df['hla.allele'] + '+' + train_df['ab']
valid_df['tag'] = valid_df['pep'] + '+' + valid_df['hla.allele'] + '+' + valid_df['ab']

immrep_df = pd.read_csv('../data/immrep2023_solutions.csv')
immrep_df = immrep_df[immrep_df['Label']==1]
immrep_df['HLA'] = immrep_df['HLA'].apply(lambda x: 'HLA-'+x)
immrep_df['ab'] = immrep_df['CDR3a'] + '/' + immrep_df['CDR3b']
immrep_df['ab_extended'] = immrep_df['CDR3a_extended'] + '/' + immrep_df['CDR3b_extended']
immrep_df['tag'] = immrep_df['Peptide'] + '+' + immrep_df['HLA'] + '+' +immrep_df['ab']
immrep_df['tag_extended'] = immrep_df['Peptide'] + '+' + immrep_df['HLA'] + '+' +immrep_df['ab_extended']
filtered_tags = list(set(immrep_df['tag'].tolist()) | set(immrep_df['tag_extended'].tolist()))

filtered_train_df = train_df[~train_df['tag'].isin(filtered_tags)]
filtered_valid_df = valid_df[~valid_df['tag'].isin(filtered_tags)]
print(f'Train positives: {len(train_df):6d} -> {len(filtered_train_df):6d} ({len(train_df)-len(filtered_train_df):4d} filtered)')
print(f'Valid positives: {len(valid_df):6d} -> {len(filtered_valid_df):6d} ({len(valid_df)-len(filtered_valid_df):4d} filtered)')

filtered_train_df = filtered_train_df.copy().drop(columns=['tag'])
filtered_valid_df = filtered_valid_df.copy().drop(columns=['tag'])
filtered_train_df.to_csv(os.path.join(data_path, "filtered_train_data_fold0.csv"), index=False)
filtered_valid_df.to_csv(os.path.join(data_path, "filtered_valid_data_fold0.csv"), index=False)
```

level 4 (no overlap, no filtering)

```python
level = "4"
data_path = f"./level{level}/"
train_df = pd.read_csv(os.path.join(data_path,"train_data_fold{}.csv".format(0)))
valid_df = pd.read_csv(os.path.join(data_path,"valid_data_fold{}.csv".format(0)))
train_df = train_df[train_df['label']==1]
valid_df = valid_df[valid_df['label']==1]

train_df['TCRab'] = train_df['ab_cdr3'] + '-' + \
    train_df['alpha.v'] + '/' + train_df['alpha.j'] + '/' + train_df['beta.v'] + '/' + train_df['beta.j']
valid_df['TCRab'] = valid_df['ab_cdr3'] + '-' + \
    valid_df['alpha.v'] + '/' + valid_df['alpha.j'] + '/' + valid_df['beta.v'] + '/' + valid_df['beta.j']

train_df['tag'] = train_df['pep'] + '+' + train_df['hla.allele'] + '+' + train_df['TCRab']
valid_df['tag'] = valid_df['pep'] + '+' + valid_df['hla.allele'] + '+' + valid_df['TCRab']

immrep_df = pd.read_csv('../data/immrep2023_solutions.csv')
immrep_df = immrep_df[immrep_df['Label']==1]
immrep_df['HLA'] = immrep_df['HLA'].apply(lambda x: 'HLA-'+x)

immrep_df['TCRab'] = immrep_df['CDR3a'] + '/' + immrep_df['CDR3b'] + '-' + \
    immrep_df['Va'] + '/' + immrep_df['Ja'] + '/' + immrep_df['Vb'] + '/' + immrep_df['Jb']
immrep_df['TCRab_extended'] = immrep_df['CDR3a_extended'] + '/' + immrep_df['CDR3b_extended'] + '-' + \
    immrep_df['Va'] + '/' + immrep_df['Ja'] + '/' + immrep_df['Vb'] + '/' + immrep_df['Jb']

immrep_df['tag'] = immrep_df['Peptide'] + '+' + immrep_df['HLA'] + '+' +immrep_df['TCRab']
immrep_df['tag_extended'] = immrep_df['Peptide'] + '+' + immrep_df['HLA'] + '+' +immrep_df['TCRab_extended']
filtered_tags = list(set(immrep_df['tag'].tolist()) | set(immrep_df['tag_extended'].tolist()))

filtered_train_df = train_df[~train_df['tag'].isin(filtered_tags)]
filtered_valid_df = valid_df[~valid_df['tag'].isin(filtered_tags)]
print(f'Train positives: {len(train_df):6d} -> {len(filtered_train_df):6d} ({len(train_df)-len(filtered_train_df):4d} filtered)')
print(f'Valid positives: {len(valid_df):6d} -> {len(filtered_valid_df):6d} ({len(valid_df)-len(filtered_valid_df):4d} filtered)')
```
