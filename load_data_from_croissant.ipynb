Load the full dataset

```python
import mlcroissant as mlc

dataset = mlc.Dataset(jsonld="croissant.json")
```

Change `record_set` to load datasets at different levels

```python
records = dataset.records(record_set="level-I")

for i, record in enumerate(records):
    print(record)
    if i > 5:
        break
```

```python
records = dataset.records(record_set="level-IIA")

for i, record in enumerate(records):
    print(record)
    if i > 5:
        break
```

```python
records = dataset.records(record_set="level-IIB")

for i, record in enumerate(records):
    print(record)
    if i > 5:
        break
```

```python
records = dataset.records(record_set="level-III")

for i, record in enumerate(records):
    print(record)
    if i > 5:
        break
```

```python
records = dataset.records(record_set="level-IV")

for i, record in enumerate(records):
    print(record)
    if i > 5:
        break
```
